```python
ax = dfc[dfc['transcripts']==28808.0].plot(x='ribosomes', y='collisions_percentage', figsize=(10, 7), fontsize=16)
dfc[dfc['transcripts']==57558.0].plot(x='ribosomes', y='collisions_percentage', ax=ax)
ax.legend(["29000 transcripts", "58000 transcripts"], prop={'size': 16})
```

```python
dftnotnan = dft[dft.tic_toc.notnull()]
del dftnotnan['index']
dftnotnan.head()
```

```python
dftnotnan_grouped_28808 = dftnotnan[dftnotnan['transcripts']==28808].groupby(['geneID', 'init_rate', 'length', 'ribosomes', 'transcripts'], 
                                                   as_index=False).mean()
dftnotnan_grouped_57558 = dftnotnan[dftnotnan['transcripts']==57558].groupby(['geneID', 'init_rate', 'length', 'ribosomes', 'transcripts'], 
                                                   as_index=False).mean()
```

```python
dftnotnan_grouped_28808.head(n=20)
```

```python
dftnotnan_grouped_57558.tail(n=20)
```

The tic-toc is getting faster as the number of ribosomes increases; this may explain the observed effect. 

Reason might be that the number of tRNAs grows proportional to the number of ribosomes (TODO: check by comparing with constant tRNAs run).
