This notebook imports the Supplementary Data from "Periodic mRNA synthesis and degradation co‐operate during cell cycle gene expression" (http://msb.embopress.org/content/10/1/717) into a Python dictionary and saves it as a `pickle` file.

```python
import os.path
import cPickle as pkl
import xlrd
```

Files and directories:

```python
datadir = "../data"
replica_files = ["inline-supplementary-material-3.xlsx", "inline-supplementary-material-4.xlsx"]
```

```python
for replica_file in replica_files:
    workbook = xlrd.open_workbook(os.path.join(datadir, replica_file))
    worksheet = workbook.sheet_by_name('Sheet1')
    num_rows = worksheet.nrows - 1
    num_cells = worksheet.ncols - 1
    curr_row = -1
    while curr_row < num_rows:
        curr_row += 1
        row = worksheet.row(curr_row)
        print 'Row:', curr_row
        curr_cell = -1
        while curr_cell < num_cells:
            curr_cell += 1
            # Cell Types: 0=Empty, 1=Text, 2=Number, 3=Date, 4=Boolean, 5=Error, 6=Blank
            cell_type = worksheet.cell_type(curr_row, curr_cell)
            cell_value = worksheet.cell_value(curr_row, curr_cell)
            #print '	', cell_type, ':', cell_value
```
