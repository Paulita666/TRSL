## Comparison and export of protein decay rates

```python
%matplotlib inline
import math
import os.path
import csv
import cPickle as pkl
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

datadir = "../data"
```

Half lives according to Belle et al. (http://www.pnas.org/content/103/35/13004.full?tab=ds) (original half lives in minutes):

```python
half_lifes_file = r"Protein_half_lifes_SuppDataSet.csv"
half_lifes_filepath = os.path.join(datadir, half_lifes_file)
```

```python
decay_constants = {}
with open(half_lifes_filepath, mode='r') as infile:
    reader = csv.reader(infile, delimiter=';')
    reader.next() # skip header line
    for rows in reader:
        decay_constants[rows[0]] = math.log(2.)/(float(rows[3])*60.) # in s^-1
```

```python
pd.Series(decay_constants).describe()
```

Correlate decay rates with ORF lengths, initiation rates, proteome, transcriptome:

```python
file_siwiak = r'journal.pcbi.1000865.s003.CSV'
filepath_siwiak = os.path.join(datadir, file_siwiak)

with open(filepath_siwiak, mode='r') as infile:
    content = [line.split(',') for line in infile.readlines()[1:]]
    init_rates_siwiak = {elem[0]: float(elem[7]) for elem in content}
    orf_lengths = {elem[0]: float(elem[1]) for elem in content} # length of the transcript CDS in codons
```

```python
init_rates_plotkin = pkl.load(open("../parameters/init_rates_plotkin.p"))
pd.Series(init_rates_plotkin).describe()
```

```python
prot_paxdb = pkl.load(open("../parameters/prot_paxdb.p"))
pd.Series(prot_paxdb).describe()
```

```python
transcriptome = pkl.load(open("../parameters/transcriptome_60000.p"))
pd.Series(transcriptome).describe()
```

```python
df1 = pd.DataFrame.from_dict(decay_constants.items())
df1.columns = ['name', 'decay_constants']

df2 = pd.DataFrame.from_dict(orf_lengths.items())
df2.columns = ['name', 'orf_lengths']

df3 = pd.DataFrame.from_dict(init_rates_plotkin.items())
df3.columns = ['name', 'init_rates_plotkin']

df4 = pd.DataFrame.from_dict(prot_paxdb.items())
df4.columns = ['name', 'prot_paxdb']

df5 = pd.DataFrame.from_dict(transcriptome.items())
df5.columns = ['name', 'transcriptome']
```

```python
df = pd.merge(df1, df2, left_on='name', right_on='name', how='outer')
df = pd.merge(df, df3, left_on='name', right_on='name', how='outer')
df = pd.merge(df, df4, left_on='name', right_on='name', how='outer')
df = pd.merge(df, df5, left_on='name', right_on='name', how='outer')
```

```python
axl = pd.scatter_matrix(df, alpha=0.2, figsize=(15, 15), diagonal='kde')

for i, axs in enumerate(axl):
    for j, ax in enumerate(axs):
        mini = max(0.0001, df[df.columns[i+1]].min())
        minj = max(0.0001, df[df.columns[j+1]].min())
        if i != j:  # only the scatter plots
            # scale plot axes by their mins and maxs
            ax.set_xlim([minj, df[df.columns[j+1]].max()]) 
            ax.set_ylim([mini, df[df.columns[i+1]].max()])
            ax.set_xscale('log')
            ax.set_yscale('log')
        else:       # only the density plots
            ax.set_xlim([minj, df[df.columns[j+1]].max()])
            ax.set_xscale('log')
```

The decay constants by Belle et al. do not seem to be correlated with anything.

Half lives according to Nagaraj et al. (http://www.cell.com/action/showImagesData?pii=S2211-1247%2814%2900934-6) (original half lives in minutes):

```python
import xlrd
```

```python
decay_nagaraj_file = '../data/Nagaraj_mmc2.xlsx'

workbook = xlrd.open_workbook(decay_nagaraj_file)
worksheet = workbook.sheet_by_name('Sheet1')
num_rows = worksheet.nrows - 1
num_cells = worksheet.ncols - 1
curr_row = 0 # First row contains headers hence not -1
decay_rates_nagaraj = {}

for curr_row in range(worksheet.nrows)[1:]:
    row = worksheet.row(curr_row)
    genename = row[1].value.split(';')[0]
    if genename:
        if isinstance(row[2].value, float):
            decay_rates_nagaraj[genename] = row[2].value * 60
```

```python
pd.Series(decay_rates_nagaraj).describe()
```

```python
plt.hist(decay_rates_nagaraj.values(), bins = 10**np.linspace(np.log10(min(decay_rates_nagaraj.values())), 
                                                              np.log10(max(decay_rates_nagaraj.values())), 50))
plt.gca().set_xscale("log")
plt.gca().set_xlim([min(decay_rates_nagaraj.values()), max(decay_rates_nagaraj.values())])
plt.gca().set_xlabel('decay constant [1/s]')
plt.show()
```

```python
rates = pd.DataFrame.from_dict(decay_constants.items())
rates.columns = ['name', 'decay_constants']

rates2 = pd.DataFrame.from_dict(decay_rates_nagaraj.items())
rates2.columns = ['name', 'decay_rates_nagaraj']

df = pd.merge(rates, rates2, left_on='name', right_on='name', how='outer')
```

```python
axl = pd.scatter_matrix(df, alpha=0.2, figsize=(5, 5), diagonal='kde')

for i, axs in enumerate(axl):
    for j, ax in enumerate(axs):
        mini = max(0.0001, df[df.columns[i+1]].min())
        minj = max(0.0001, df[df.columns[j+1]].min())
        if i != j:  # only the scatter plots
            # scale plot axes by their mins and maxs
            ax.set_xlim([minj, df[df.columns[j+1]].max()]) 
            ax.set_ylim([mini, df[df.columns[i+1]].max()])
            ax.set_xscale('log')
            ax.set_yscale('log')
        else:       # only the density plots
            ax.set_xlim([minj, df[df.columns[j+1]].max()])
            ax.set_xscale('log')
```

No correlation between the decay rates by Belle and Nagaraj!
